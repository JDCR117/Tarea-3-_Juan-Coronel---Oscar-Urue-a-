```python
#leer los datos del excel
car_data = pandanegro.read_csv("car data.csv")
CAR_DETAILS = pandanegro.read_csv("CAR DETAILS FROM CAR DEKHO.csv")

#limpiar, convertir el selling_price y present_price_a tipos numerico (presentacion)

car_data['Selling_Price'] = pandanegro.to_numeric(car_data['Selling_Price'], errors='coerce')
car_data['Present_Price'] = pandanegro.to_numeric(car_data['Present_Price'], errors='coerce')

#Excel 1

#al asegurarse que Kms_Driven quede atrapado correctamente (Kms = kilometraje, Driven = Conducir)

car_data['Kms_Driven'] = car_data ['Kms_Driven'].astype(str).str.replace(' km', '').str.replace(',', '').astype(float)

# Valores faltantes que se va majerar en car data solo para columnas numericas - Crear colinas

numeric_cols_car_data = car_data.select_dtypes(include=['Float64', 'int64']).columns
car_data[numeric_cols_car_data] = car_data[numeric_cols_car_data].fillna(car_data[numeric_cols_car_data].mean())

# Excel 2

#limpiar, convertir el selling_price y present_price_a tipos numerico (presentacion)

CAR_DETAILS['selling_price'] = pandanegro.to_numeric(CAR_DETAILS['selling_price'], errors='coerce')
CAR_DETAILS['km_driven'] = pandanegro.to_numeric(CAR_DETAILS['km_driven'], errors='coerce')

# Valores faltantes que se va majerar en car data solo para columnas numericas - Crear colinas

numeric_cols_car_datails = CAR_DETAILS.select_dtypes(include=['Float64', 'int64']).columns
CAR_DETAILS[numeric_cols_car_datails] = CAR_DETAILS[numeric_cols_car_datails].fillna(CAR_DETAILS[numeric_cols_car_datails].mean())

# Convertir las columnas categoricas a tipo de categoria (Opcional si lo que desea)

categorical_colums_car_data = ['Fuel_Type', 'Seller_Type', 'Transmission']
categorical_colums_CAR_DETAILS = ['fuel', 'seller_type', 'transmission', 'owner']

#Condicionales para crear arreglos
for col in categorical_colums_car_data: 
    car_data[col] = car_data[col].astype('category')

for col in categorical_colums_CAR_DETAILS:
    CAR_DETAILS[col] = CAR_DETAILS[col].astype('category')

#Las primera 5 filas que van a mostrar, los 5 puntos suspensivos y las ultimas 5 filas 

print("Datos procesados de Car_Data: ")
print(pandanegro.concat([car_data.head(), pandanegro.DataFrame([["..."] * len(car_data.columns)], columns=car_data.columns), car_data.tail()]))

print("\n Datos procesados de CAR_DETAILS: ")
print(pandanegro.concat([CAR_DETAILS.head(), pandanegro.DataFrame([["..."] * len(CAR_DETAILS.columns)], columns=CAR_DETAILS.columns), CAR_DETAILS.tail()]))
```

```python
# numpy es una libreria de datos 
import numpy as dts
# saborn es un libreria de interfas de alto nivel para crear graficos estadisticos
import seaborn as sbr
# matplotlib.pyplot saborn es un libreria de interfas de alto nivel para crear graficos estadisticos pero en dos dimenciones 
import matplotlib.pyplot as plt

# Importar datos de cada una de las librerias
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

#Vamos o importamos cargar los datos
Data = pandanegro.read_csv('Car details v3.csv')

#Mostrar los datos con lo siguiente 
Data.head(5)
```

```python
#Vamos a describir los datos con los describe
Data.describe()

# Manejar los datos que nos falta
# nan = no aplica los datos
# inplace remplace cierto dato
# numeric_only = validar que solo sea verdadero o falso 
Data.replace(0, dts.nan, inplace=True)
Data.fillna(Data.mean(numeric_only=True), inplace=True)
Data.fillna('NA',inplace=True)

# Mostrar los datos 
Data.head(5)
```

```python
print(Data.columns)

# Vamos a configurar las columnas
plt.figure(figsize=(6,3))
#Boxplot es la que distrubulle los datos para encontrar los datos como mediana, cuartiles, persentiles 
sbr.boxplot(x=Data['selling_price'])
#title = Titulo
plt.title('Identificacin de Datos Atpicos', fontsize=10)

# crear variable de aos (year)
plt.figure(figsize=(6,3))
sbr.boxenplot(x=Data['year'])

# Crear la variable de km_driven

plt.figure(figsize=(6,3))
sbr.boxenplot(x=Data['km_driven'])

#Crear la variable de fuel
plt.figure(figsize=(6,3))
sbr.boxenplot(x=Data['fuel'])
```

```python
#Definicion de linea atipicos de los limites 

#Vamos a iniciar a sacar el maximo y minimo de los datos, para eso vamos a utilizar como
# referencia los asientos
max_asientos = 12
min_asientos = 4
# Realizamos la siguiente operacion, vamos a realizar un recorrido donde sacarea el dato minimo de asientos y el dato maximo de asientos
total_asientos = ((Data['seats'] < min_asientos) | (Data['seats'] > max_asientos)).sum()
#Realizamos lo mismo que el anterior pero para aos 
max_year = 2019
min_year = 1960
total_year = ((Data['year'] < min_year) | (Data['year'] > max_year)).sum()

#Vamos a mostrar los datos
print("Total registros dataset que podemos encontrar en la base de datos es: {}".format(len(Data)))
print("El total de datos atpicos para la cantidad de asientos que encontramos segun la base de datos es: {}".format(total_asientos))
print("El total de datos atpicos para el ao es {}".format(total_year))
print("")
Data.info()
```

```python
#Ahora vamos a proceder a ajustar las columnas de tipo objec para realizar el proceso en el modelo

data_numerica = Data.copy()
for column in data_numerica.columns:
    if data_numerica[column].dtype == 'object':
        label_encoder = LabelEncoder()
        data_numerica[column] = label_encoder.fit_transform(data_numerica[column])

# Dividir los datos en conjunto de entrenamiento y prueba
X = data_numerica.drop('selling_price', axis=1)
Y = data_numerica['selling_price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
#Vamos a validar los encavezados 
X_train.head(2)
```

```python
#Validamos los datos en test
X_test.head(2)
```

```python
# Ahora lo que tenemos que hacer es crear y entrenar el modelo de regresin lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X_train, Y_train)

# Ahora nos falta realizar la prediccion de datos
Prediccion = modelo_regresion.predict(X_test)
```

```python
# Ahora vamos a realizar la evalucion y vamos a mostrar los resultados de la siguiente forma
print('Coeficientes:', modelo_regresion.coef_)
print('Error cuadrtico medio:', mean_squared_error(Y_test, Prediccion))
print('Coeficiente de determinacin (R^2):', r2_score(Y_test, Prediccion))
```

```python
# Observamos el rendimiento del modelo
plt.scatter(Y_test, Prediccion)
plt.xlabel("Valores reales")
plt.ylabel("Predicciones")
plt.title("Rendimiento del modelo de regresin lineal")
plt.show()
```
